```python
%config IPCompleter.greedy=True
import pandas as pd
import missingno as mn
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
import math
pd.set_option('display.max_columns', None)
```

```python
score_base = 1
epsilon = 0.01
weights = { 'p' : 0.35, 'd': 0.35, 'o': 0.1}

def read_request_data():
    request_data = pd.read_csv('./Data/search_requests.csv', sep=',', low_memory=False)
    request_data.replace({"\\\\N": np.nan}, inplace=True)
    request_data["SEARCH_TIMESTAMP"] = request_data["SEARCH_TIMESTAMP"].astype("datetime64")
    request_data["SEARCH_DURATION_IN_DAYS"] = request_data["SEARCH_DURATION_IN_DAYS"].astype("int")
    request_data["USER_NB_BOOKINGS"] = request_data["USER_NB_BOOKINGS"].astype("float")
    request_data["USER_NB_OPEN_BOOKINGS"] = request_data["USER_NB_OPEN_BOOKINGS"].astype("float")
    request_data["USER_NB_INSTANT_BOOKINGS"] = request_data["USER_NB_INSTANT_BOOKINGS"].astype("float")
    request_data["USER_NB_CITY_BOOKINGS"] = request_data["USER_NB_CITY_BOOKINGS"].astype("float")
    request_data["USER_NB_FAMILY_BOOKINGS"] = request_data["USER_NB_FAMILY_BOOKINGS"].astype("float")
    request_data["USER_NB_COMMERCIAL_VAN_BOOKINGS"] = request_data["USER_NB_COMMERCIAL_VAN_BOOKINGS"].astype("float")
    request_data["USER_NB_SEDAN_BOOKINGS"] = request_data["USER_NB_SEDAN_BOOKINGS"].astype("float")
    request_data["USER_AVG_BOOKING_PRICE"] = request_data["USER_AVG_BOOKING_PRICE"].astype("float")
    request_data["USER_AVG_BOOKING_DURATION"] = request_data["USER_AVG_BOOKING_DURATION"].astype("float")
    request_data["FILTER_PRICE_MAX"] = request_data["FILTER_PRICE_MAX"].astype("float")
    #request_data.FILTER_TRANSMISSION.replace({"[\"automatic\"]": "automatic"}, inplace=True)
    return request_data

def read_cars_data():
    path = r'./Data/impressions/' 
    all_files = glob.glob(path + "/*.csv")
    li = []
    for filename in all_files:
        df = pd.read_csv(filename, index_col=None, header=0, sep=',', low_memory=False)
        li.append(df)
    cars_data = pd.concat(li, axis=0, ignore_index=True)
    cars_data.CAR_OWNER_PRR.replace({"\\\\N": 0.5}, inplace=True)
    return cars_data

def generate_reduced_data(request_id, data):
    return data[data.SEARCH_REQUEST_ID == request_id]

def get_distance_score(data, car_id):
    #distance score:
    column = 'CAR_DISTANCE_TO_SEARCH_ADDRESS'
    max_value = data.loc[data[column].idxmax()][column]
    min_value = data.loc[data[column].idxmin()][column]
    x_value = data.loc[data['CAR_ID'] == car_id][column].tolist()[0]
    factor_2 = max_value - x_value
    if factor_2 > 0:
        return score_base / ((max_value - min_value) / factor_2) + epsilon
    else:
        return epsilon

def get_owner_prr(cars_data, car_id):
    column = 'CAR_OWNER_PRR'
    return float(cars_data.loc[cars_data['CAR_ID'] == car_id][column].tolist()[0]) + epsilon
    
def get_price_score(cars_data, car_id):
    column = 'CAR_SEARCH_PRICE'
    max_value = cars_data.loc[cars_data[column].idxmax()][column]
    min_value = cars_data.loc[cars_data[column].idxmin()][column]
    x_value = cars_data.loc[cars_data['CAR_ID'] == car_id][column].tolist()[0]
    factor_2 = max_value - x_value
    if factor_2 > 0:
        return score_base / ((max_value - min_value) / factor_2) + epsilon
    else:
        return epsilon
    
    
def get_price_bonus_score(requests_data, cars_data, car_id):
    try:
        usr_avg_price = requests_data['USER_AVG_BOOKING_PRICE'].tolist()[0]
        usr_avg_duration = requests_data['USER_AVG_BOOKING_DURATION'].tolist()[0]
        if (type(usr_avg_price) is str or not math.isnan(usr_avg_price)) and (type(usr_avg_duration) is str or not math.isnan(usr_avg_duration)):
            column = 'CAR_SEARCH_PRICE'
            usr_avg_price = float(usr_avg_price)
            usr_avg_duration = float(usr_avg_duration)
            current_duration = cars_data.loc[cars_data['CAR_ID'] == car_id][column].tolist()[0]
            estimated_price = (usr_avg_price/usr_avg_duration) * (requests_data['SEARCH_DURATION_IN_DAYS'].tolist()[0])
            #print("estimated_price ",estimated_price)
            min_proposed_price = cars_data.loc[cars_data[column].idxmin()][column]
            #print("min ",min_proposed_price)
            max_proposed_price = cars_data.loc[cars_data[column].idxmax()][column]
            #print("max ",max_proposed_price)
            current_car_price = cars_data.loc[cars_data['CAR_ID'] == car_id][column].tolist()[0]
            #print("x ",current_car_price)
            if current_car_price > estimated_price:
                return 0
                
            factor_2 = estimated_price - current_car_price
            if factor_2 > 0:
                return score_base / ((estimated_price - min_proposed_price) / factor_2)
        else:
            return 0
                
    except Exception as e: 
        print(str(e))
        return 0
    

def get_price_extra_score(data, car_id):
    column = 'CAR_PRICE_EXTRA_KM'
    max_value = data.loc[data[column].idxmax()][column]
    min_value = data.loc[data[column].idxmin()][column]
    x_value = data.loc[data['CAR_ID'] == car_id][column].tolist()[0]
    factor_2 = max_value - x_value
    if factor_2 > 0:
        return score_base / ((max_value - min_value) / factor_2) + epsilon
    else:
        return epsilon
    
def get_car_age_score(data, car_id):
    column = 'CAR_RELEASE_YEAR'
    max_value = data.loc[data[column].idxmax()][column]
    min_value = data.loc[data[column].idxmin()][column]
    x_value = data.loc[data['CAR_ID'] == car_id][column].tolist()[0]
    factor_2 = x_value - min_value
    if factor_2 > 0:
        return score_base / ((max_value - min_value) / factor_2) + epsilon
    else:
        return epsilon
    
def get_user_preferences(search_request_id, data):
    try:
        result = dict()
        nb_all_bookings = int(data['USER_NB_BOOKINGS'].tolist()[0])
        nb_open_bookings = int(data['USER_NB_OPEN_BOOKINGS'].tolist()[0])    
        nb_instant_bookings = int(data['USER_NB_INSTANT_BOOKINGS'].tolist()[0]) 
        result['OPEN_TRUE'] = nb_open_bookings / nb_all_bookings
        result['OPEN_FALSE'] = 1 - result['OPEN_TRUE']
        result['INST_TRUE'] = nb_instant_bookings / nb_all_bookings
        result['INST_FALSE'] = 1 - result['INST_TRUE']
        return result
    except:
        return None

def get_car_open_pref_score(reduced_cars_data, car_id, search_request_id, reduced_request_data):
    user_pref = get_user_preferences(search_request_id, reduced_request_data)
    if user_pref:
        data_row = reduced_cars_data[reduced_cars_data.CAR_ID == car_id]
        current_car_open_specs = data_row['CAR_IS_OPEN'].tolist()[0]
        if current_car_open_specs:
            open_pref = user_pref['OPEN_TRUE'] * score_base + epsilon
        else:
            open_pref = user_pref['OPEN_FALSE'] * score_base + epsilon
            
        current_car_inst_specs = data_row['CAR_INSTANT_BOOKABLE'].tolist()[0]
        if current_car_inst_specs:
            inst_pref = user_pref['INST_TRUE'] *  score_base + epsilon
        else:
            inst_pref = user_pref['INST_FALSE'] * score_base + epsilon
        
        return (open_pref, inst_pref)
    else:
        return None, None
    
        
def get_score(search_request_id, cars_data, request_data):
    reduced_cars_data = generate_reduced_data(search_request_id, cars_data)
    reduced_request_data = generate_reduced_data(search_request_id, request_data)
    list_of_cars = reduced_cars_data.CAR_ID.tolist()
    cars_scores = []
    for car in list_of_cars:
        
        all_scores = np.array([])
        all_scores = np.append(all_scores, weights['o']* get_car_age_score(reduced_cars_data,car))
        all_scores = np.append(all_scores, weights['p']*((get_price_score(reduced_cars_data,car) + get_price_extra_score(reduced_cars_data,car)) + get_price_bonus_score(reduced_request_data, reduced_cars_data, car)/3))    
        all_scores = np.append(all_scores, weights['d']*get_distance_score(reduced_cars_data,car))   
        
        open_pref, inst_pref = get_car_open_pref_score(reduced_cars_data, car, search_request_id, reduced_request_data)
        if open_pref and inst_pref:
            all_scores =  np.append(all_scores, weights['o'] * open_pref)
            all_scores =  np.append(all_scores, weights['o'] * inst_pref)
        prr = get_owner_prr(reduced_cars_data,car)    
        mean = np.mean(all_scores)
        mean *= prr
        cars_scores.append(mean)
    return pd.DataFrame({'CAR_ID': list_of_cars, 'CAR_SCORE': cars_scores})

def create_average_price_feature(requests_data, cars_data, request_id):
    duration = requests_data[requests_data.SEARCH_REQUEST_ID == request_id]['SEARCH_DURATION_IN_DAYS'].tolist()[0]
    list_of_cars = cars_data.CAR_ID.tolist()
    cars_scores=[]
    for car in list_of_cars:
        if duration > 0:
            avg_price = float(cars_data[cars_data.CAR_ID == car]['CAR_SEARCH_PRICE'].tolist()[0]) / duration
        else:
            avg_price = 0
        cars_scores.append(avg_price)
    return pd.DataFrame({'CAR_ID': list_of_cars, 'CAR_AVERAGE_PRICE_PER_DAY': cars_scores})
```

To start our feature engineering, let's take a subset of data that represents only one request.

```python
request_id = '0b0bbba7f09a2eb8c6cdba0e8b85e85b'
red_cars =  generate_reduced_data(request_id, cars_full_data)
red_reqs =  generate_reduced_data(request_id, request_full_data)
```

Explaining the scoring system:
From analysing the data, we notice a feature which represents the current ranking system based on current score. The task being not to replicate the current ranking, this system cannot be used as a ground-truth model.  Any supervied models cannot be applied on this given column.

The process i used in order to create my scoring model is by following these steps:
   - Ask colleagues and friends to rank based on the importance of factors in renting a car, these being : price, distance, The car model, car age
   - Try Drivy App in real life situations in order to analyse the current behaviour.
   - Check online (Kaggle, Quora mostly) for similar Problems/Models and analyse the proposed approach; a similar issue being 'Renting an apartement for holidays'
   
My proposed approch works as follow:

Let $ S(car\_id) $ be a function that return a probablibity score between 0 and 1. Which is definied as follow :

$$ S(car\_id) = ( \frac{1}{n} \sum_{\substack{   0<i<n  }} w_i * P(x_i) ) * R(car\_id)  $$

For each feature $x_i \in \{ car\_price, car\_distance, car\_age, user\_preference\_open, user\_preference\_inst \}  $

Where  : 

$ w_i $ Weight of $x_i$ that represents its importance (p: price, d: distance, o:others)

$ R(car\_id) $ : the car's owner acceptance rate

Let's dive deeper for each $x_i$ :

**1-  Car price :**
The probability of a car being selected based on price only is calculated based on three parameters:
<li>$ P_{car\_price_1}(car\_id) $ The car's price based on the request: with $0+ \epsilon$ being the probability of choosing the car with the highest price, $1$ for chosing the car with the lowest price</li>
<li>$ P_{car\_price_2}(car\_id) $ The car's extra price : with $0+ \epsilon$ being the probability of choosing the car with the highest price, $1$ for chosing the car with the lowest price</li>
<li>$ P_{car\_price_3}(car\_id) $ The current offer compared to the usual renting price by the user, with $0+ \epsilon$ attributed to the offers that are either higher or equal to the usual renting price, $1$ for the car with the best offer compared to the usual price </li>
The score is is calculated as follow:
$$ P(x_i = car\_price) = \frac{1}{3} \sum_{\substack{   0<i<=3  }}  P_{car\_price_i}(car\_id)  $$

**2-  Car distance / Car age:**
The probability of a car being selected based on distance only or age only is calculated as follow:
$$ P(x_i = car\_distance) = 
  \begin{cases}
    0+ \epsilon       & \quad \text{Furthest / oldest car in the pool}\\
    1       & \quad \text{Closest / newest car in the pool}
  \end{cases}
$$


**3-  User preferences:**
This attribute is calculated as follow:
<li> Probability of the the car being selected based on its OPEN value, which is calculated as following </li>
$$ P(x_i = user\_pref\_open) = 
  \begin{cases}
    P(x_i / V(car\_opened)=true )       & \quad \text{If V(car_opened)=true : return pourcentage of opened car booked by the user}\\
    1-P(x_i / V(car\_opened)=true )        & \quad \text{If V(car_opened)=false}
  \end{cases}
$$

<li> Probability of the the car being selected based on its INSTANT_BOOKABLE value, which is calculated as following </li>
$$ P(x_i = user\_pref\_inst) = 
  \begin{cases}
    P(x_i / V(car\_inst)=true )       & \quad \text{If V(car_inst)=true : return pourcentage of intant bookable cars booked by the user}\\
    1-P(x_i / V(car\_inst)=true )        & \quad \text{If V(car_inst)=false}
  \end{cases}
$$

<a id="ref3"><h2>Critics and future ideas</h2></a>
#### Critics 
The challenge with working on this particular scoring model, is that there isn't a ground truth model upon which we can evaluate our results.
The scoring model proposed has some flaws which are:
<li> Considering that features are independant. For example, based on the correlation graph, $cov(car\_is\_open, car\_instant\_bookable) = 0.870173 $  which means that these features are highly corelated. Correlation and causality should be checked before calculating the probabilty.</li>
<li> Weights $w_i$ are chosen based on a small survey,  these value can bias the scoring model and should be reconsidered</li>


#### Future ideas


<li> Create user profiles where each profile is defined by a set of renting habbits, criterias importance (prince, distance...)  </li> 
<li> Capture what the user is really looking for, based on his history or/and using the behaviour of it profiling data </li> 
<li> Create a score of each car model, to try to propose cars that are highly demanded </li> 
<li> Utilize cars ratings, and use NLP methods on user's feedback to improve the rating </li>
<li> Create a score on car owners and prioritize well rated owners </li>
